```python
#PRINTING MONTHLY GROWTH RATES
from user_inputs import growth_rate_dict, year_start, baseline_suspended_sediment
from config import  sediment_years_months
from utils import calculate_monthly_adjusted_growth_rate

print("Year\tMonth\tMorphology\tMonthly Growth Rate (cm/month)")
for (year_offset, month), values in sediment_years_months.items():
    suspended, _ = values
    year = year_start + year_offset
    additional_suspended = suspended - baseline_suspended_sediment
    if additional_suspended < 0:
        additional_suspended = 0
    for morphology in growth_rate_dict.keys():
        annual_rate = growth_rate_dict[morphology]
        monthly_rate = calculate_monthly_adjusted_growth_rate(
            morphology, annual_rate, additional_suspended
        )
        print(f"{year}\t{month}\t{morphology}\t{monthly_rate:.3f}")
```

```python
# Model Outputs Summary

import pandas as pd
import matplotlib.pyplot as plt
from config import calculate_additional_sediment, growth_rate, coral_type, year_start
from utils import *
import numpy as np

# --- 1. Additional Sediment Through Time ---
additional_sediment = calculate_additional_sediment()
sediment_df = pd.DataFrame([
    {'Year': year, 'Month': month, 'Additional Suspended': vals[0], 'Additional Deposited': vals[1]}
    for (year, month), vals in additional_sediment.items()
])
sediment_yearly = sediment_df.groupby('Year')[['Additional Suspended', 'Additional Deposited']].sum().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(sediment_yearly['Year'], sediment_yearly['Additional Suspended'], label='Suspended Sediment')
plt.plot(sediment_yearly['Year'], sediment_yearly['Additional Deposited'], label='Deposited Sediment')
plt.xlabel('Year')
plt.ylabel('Additional Sediment')
plt.title('Additional Sediment Through Time')
plt.legend()
plt.tight_layout()
plt.show()

# --- 2. Growth Rates Through Time (per morphology group) ---
plt.figure(figsize=(10, 6))
for morph in coral_type:
    yearly_growth = growth_rate[morph].groupby('Year').mean()
    plt.plot(yearly_growth.index, yearly_growth.values, label=morph)
plt.xlabel('Year')
plt.ylabel('Mean Growth Rate')
plt.title('Growth Rates Through Time by Morphology Group')
plt.legend()
plt.tight_layout()
plt.show()

# --- 3. Area of Coral Morphology Groups Through Time ---
# Assuming you have a DataFrame 'final_df' with columns: 'Year', 'Branching_Area (%)', 'Foliose_Area (%)', 'Other_Area (%)'
# If not, you can generate it using your model run functions.
try:
    # If final_df exists
    plt.figure(figsize=(10, 6))
    for morph, color in zip(['Branching_Area (%)', 'Foliose_Area (%)', 'Other_Area (%)'], ['blue', 'fuchsia', 'orange']):
        plt.plot(final_df['Year'] + year_start, final_df[morph], label=morph.replace('_Area (%)', ''), color=color)
    plt.xlabel('Year')
    plt.ylabel('Area (%)')
    plt.title('Area of Coral Morphology Groups Through Time')
    plt.legend()
    plt.tight_layout()
    plt.show()
except NameError:
    print("final_df not found. Please run the model to generate final_df.")

# --- End of Summary ---
```
