```python
#print
from user_inputs import sediment_years, year_start

def print_sediment_data():
    print("Sediment Data:")
    print("Year\tMonth\tSuspended Sediment (mg/L)\tDeposited Sediment (mg/cm²)")
    for year_offset, data in sediment_years.items():
        year = year_start + year_offset
        month, suspended_sediment, deposited_sediment = data
        print(f"{year}\t{month}\t{suspended_sediment}\t\t\t{deposited_sediment}")

# Call the function to print the sediment data
print_sediment_data()
```
