```python
#PRINTING ADDITIONAL SEDIMENT
from config import calculate_additional_sediment

additional_sediment = calculate_additional_sediment()

print("Year\tMonth\tAdditional Suspended (mg/L)\tAdditional Deposited (mg/cm²)")
for (year, month), values in additional_sediment.items():
    add_suspended, add_deposited = values
    print(f"{year}\t{month}\t{add_suspended}\t\t\t{add_deposited}")
```

```python
#PRINTING MONTHLY GROWTH RATES
from user_inputs import growth_rate_dict, year_start, baseline_suspended_sediment
from config import  sediment_years_months
from utils import calculate_monthly_adjusted_growth_rate

print("Year\tMonth\tMorphology\tMonthly Growth Rate (cm/month)")
for (year_offset, month), values in sediment_years_months.items():
    suspended, _ = values
    year = year_start + year_offset
    additional_suspended = suspended - baseline_suspended_sediment
    if additional_suspended < 0:
        additional_suspended = 0
    for morphology in growth_rate_dict.keys():
        annual_rate = growth_rate_dict[morphology]
        monthly_rate = calculate_monthly_adjusted_growth_rate(
            morphology, annual_rate, additional_suspended
        )
        print(f"{year}\t{month}\t{morphology}\t{monthly_rate:.3f}")
```

```python
#printing annual growth rates for each morphology
from user_inputs import growth_rate_dict, enable_sediment_exposure
from config import calculate_additional_sediment, calculate_annual_growth_rate

additional_sediment = calculate_additional_sediment()

for morphology, annual_growth_rate in growth_rate_dict.items():
    annual_g_rates = calculate_annual_growth_rate(morphology, annual_growth_rate, additional_sediment)
    print(f"Annual growth rates for {morphology}:")
    print("Year\tTotal Annual Growth Rate (cm/year)")
    for year, total_rate in sorted(annual_g_rates.items()):
        print(f"{year}\t{total_rate:.3f}")
    print()
```
